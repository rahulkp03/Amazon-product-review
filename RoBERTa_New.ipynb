```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('/Users/rahulkp/Downloads/amazon_reviews_us_Shoes_v1_00.tsv',sep='\t',error_bad_lines=False)
data.head()
```

```python
data.shape
```

```python
data['star_rating'].value_counts()
```

```python
data['star_rating'].value_counts()/data.shape[0]
```

```python
data['star_rating']=data['star_rating'].replace([1,2],0)
```

```python
data['star_rating']=data['star_rating'].replace([3,4,5],1)
```

```python
data['star_rating'].value_counts()/data.shape[0]
```

```python
df=data[['customer_id','product_id','review_id','review_date','review_headline','review_body','star_rating']].sample(750,random_state=50)
df
```

```python
df['star_rating'].value_counts()/df.shape[0]
```

```python
import transformers as ppb
model_class, tokenizer_class, pretrained_weights = (ppb.RobertaModel, ppb.RobertaTokenizer, 'roberta-base')
# Load pretrained model/tokenizer
tokenizer = tokenizer_class.from_pretrained(pretrained_weights)
model = model_class.from_pretrained(pretrained_weights)
```

### Tokenization

```python
tokenized = df['review_body'].apply((lambda x: tokenizer.encode(x, add_special_tokens=True,truncation=True, max_length=512)))
```

```python
max_len = 0
for i in tokenized.values:
    if len(i) > max_len:
        max_len = len(i)

padded = np.array([i + [0]*(max_len-len(i)) for i in tokenized.values])
```

```python
np.array(padded).shape
```

```python
attention_mask = np.where(padded != 0, 1, 0)
attention_mask.shape
```

```python
import torch
input_ids = torch.tensor(padded)  
attention_mask = torch.tensor(attention_mask)

with torch.no_grad():
    last_hidden_states = model(input_ids, attention_mask=attention_mask)
```

```python
features = last_hidden_states[0][:,0,:].numpy()
```

```python
labels = df['star_rating']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

1.Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr_clf = LogisticRegression()
lr_clf.fit(X_train, y_train)
```

```python
accuracy = lr_clf.score(X_test, y_test)
print("Accuracy: {:.2f}%".format(accuracy*100))
```

```python
y_pred=lr_clf.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test, (y_pred)))
print(confusion_matrix(y_test, (y_pred)))
```

```python
np.quantile(lr_clf.predict_proba(X_test)[:, 1],0.12)
```

```python
y_prob=lr_clf.predict_proba(X_test)[:,1]>0.8070140134828888
y_prob
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test, (y_prob)))
print(confusion_matrix(y_test, (y_prob)))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score
probs = lr_clf.predict_proba(X_test)[:, 1]

fpr, tpr, thresholds = roc_curve(y_test,probs)
auc = roc_auc_score(y_test, probs)

plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % auc)
plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```
